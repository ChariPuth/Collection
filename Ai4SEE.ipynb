```python
import pandas as pd
correlation_matrix = pd.read_csv('correlation_matrix.csv', index_col=0)

thresh = 0.9
strong_corr = {}

for column in correlation_matrix.columns:
    strong_corr = correlation_matrix.index[correlation_matrix[column] > thresh].tolist()
    if len(strong_corr) > 1:
       strong_corr[column] = strong_corr

for column in correlation_matrix.columns:
    strong_corr = correlation_matrix.index[correlation_matrix[column] < -thresh].tolist()
    if len(strong_corr) > 1:
        if column in strong_corr:
           strong_corr[column].extend(strong_corr)
            # Remove duplicates
           strong_corr[column] = list(set(strong_corr[column]))
        else:
           strong_corr[column] = strong_corr

# Remove self-correlations
for key, value in strong_corr.items():
   strong_corr[key] = list(set(value) - {key})


print("Variables with correlation coefficients close to 1 or -1:")
for key, value in strong_corr.items():
    print(f"{key}: {value}")
```

## d) Find out the variables which are the root cause for the anomaly

```python
import pandas as pd
from sklearn.ensemble import IsolationForest

test_data = pd.read_csv('/content/test.csv')

isolation_forest = IsolationForest(contamination=0.1)
isolation_forest.fit(test_data)

test_data['anomaly_prediction'] = isolation_forest.predict(test_data)

anomaly_features = test_data[test_data['anomaly_prediction'] == -1].drop('anomaly_prediction', axis=1)
anomaly_means = anomaly_features.mean()

most_anomalous_variables = anomaly_means.nlargest(8)
print("Variables most responsible for anomalies:")
print(most_anomalous_variables)
```
